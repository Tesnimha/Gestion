```python
import pandas as pd
import random

data_set_250 = data_set.head(250)

# Liste de médicaments fictifs
medicaments = data_set_250['brand_name'].unique().tolist()

# Créer un DataFrame vide pour les interactions médicamenteuses
df_interactions = pd.DataFrame(columns=["Médicament 1", "Médicament 2", "Interaction"])

# Générer des interactions médicamenteuses fictives
for i in range(100):  # Générer 100 interactions fictives
    medicament1 = random.choice(medicaments)
    medicament2 = random.choice(medicaments)
    interaction = random.choice(["Aucune interaction", "Interaction mineure", "Interaction majeure"])
    
    # Ajouter l'interaction fictive au DataFrame
    df_interactions = df_interactions.append({"Médicament 1": medicament1, "Médicament 2": medicament2, "Interaction": interaction}, ignore_index=True)

# Afficher le DataFrame d'interactions médicamenteuses fictives
print(df_interactions)
```
